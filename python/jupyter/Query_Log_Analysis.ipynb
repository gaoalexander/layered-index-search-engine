```python
import heapq
```

```python
query_log_path = '../../data/3_log/query_log.txt'
tier1_queries_path = '../../data/3_log/tier1_queries.txt'
tier2_queries_path = '../../data/3_log/tier2_queries.txt'
```

```python
# ONLY COUNT UNIQUE QUERIES

tier1 = set()
tier2 = set()

with open(query_log_path) as logfile:
    line = logfile.readline()
    while line:
#         terms = line.split(" ")
#         str1 = terms[1] + " " + terms[2][0:len(terms[2]) - 1]
#         str2 = terms[2][0:len(terms[2]) - 1] + " " + terms[1]
        if line[0] == '1':
            tier1.add(line[2:])
        elif line[0] == '2':
            tier2.add(line[2:])
        line = logfile.readline()
        
with open(tier1_queries_path, "w") as tier1_outfile:
    for each in tier1:
        tier1_outfile.write(each)
    
with open(tier2_queries_path, "w") as tier2_outfile:
        for each in tier2:
            tier2_outfile.write(each)

!wc -l '../../data/3_log/query_log.txt'
!wc -l '../../data/3_log/tier1_queries.txt'
!wc -l '../../data/3_log/tier2_queries.txt'
```

```python
# INCLUDE DUPLICATE QUERIES IN COUNTS

with open(query_log_path) as logfile:
    with open(tier1_queries_path, "w") as tier1_outfile:
        with open(tier2_queries_path, "w") as tier2_outfile:
            line = logfile.readline()
            while line:
                if line[0] == '1':
                    tier1_outfile.write(line[2:])
                else:
                    tier2_outfile.write(line[2:])
                line = logfile.readline()
```
